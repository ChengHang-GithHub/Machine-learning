```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def entropy(px):
    return - (px * np.log2(px))
x = np.linspace(0.01, 1, 100)#返回在0.01-1之间等间隔采样的100个点
plt.figure(figsize=(5, 3), dpi=100)#图像设置：图像大小为5*3，分辨率为200
plt.title('$Entropy(x) = - P(x) * log_2(P(x))$')
plt.xlim(0, 1)#x轴范围
plt.ylim(0, 0.6)#y轴范围
plt.xlabel('P(x)')#x轴的标签
plt.ylabel('Entropy')#y轴的标签
plt.plot(x, entropy(x), 'r-')#画图操作，指定x轴的值，指定y轴的值，以及颜色
plt.show()
```

```python
def gini_impurity(px):
    return px * (1 - px)

x = np.linspace(0.01, 1, 100)
plt.figure(figsize=(5, 3), dpi=100)
plt.title('$Gini(x) = P(x) (1 - P(x))$')
plt.xlim(0, 1)
plt.ylim(0, 0.6)
plt.xlabel('P(x)')
plt.ylabel('Gini Impurity')
plt.plot(x, entropy(x), 'r-')
plt.show()
```
